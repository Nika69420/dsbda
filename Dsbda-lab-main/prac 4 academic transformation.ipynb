# Using z-score and normalizing data

```python
from scipy.stats import zscore
def zsc(val):
    z = np.abs(zscore(val))
    return z
```

```python
df2['Age'] = zsc(df2['Age'])
df2['Marks'] = zsc(df2['Marks'])
df2
```
